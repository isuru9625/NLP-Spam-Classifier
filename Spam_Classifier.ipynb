```python
#store the sms message data
txt_msgs = df[1]
print(txt_msgs.head())
```

```python
#use regular expression to replace phone numbers, emails, urls, currencies, etc.
```

```python
processed = txt_msgs.str.replace(r'^[a-zA-Z0-9+_.-]+@[a-zA-Z0-9.-]+$'," emailaddr ")
```

```python
print(processed)
```

```python
#replace url woth webaddress
processed = processed.str.replace(r'^https?:\/\/.*[\r\n]*'," webaddr ")
```

```python
#replace money symbols with moneysymb 
processed = processed.str.replace(r'£|\$', " moneysymb ")
```

```python
#replace 10 digit phone numbers with phonenumber
processed = processed.str.replace(r'(\d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})'," phonenumber ")
```

```python
#replace normal numbers with number
processed = processed.str.replace(r'\d+(\.\d*)?'," number ")
```

```python
#remove punctuation
processed = processed.str.replace(r'[^\w\d\s]',' ')
```

```python
#replace white spaces with a single space
processed= processed.str.replace(r'\s+'," ")
```

```python
#remove leading an trailing white spaces
processed = processed.str.replace(r'^\s+|\s+?$',"")
```

```python
#change words to lowercase
processed = processed.str.lower()
```

```python
print(processed)
```

```python
nltk.download('stopwords')
```

```python
#remove stopwords
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
print(stop_words)
```

```python
processed = processed.apply(lambda x: " ".join(term for term in x.split() if term not in stop_words))
```

```python
processed.head()
```

```python
ps = nltk.PorterStemmer()
```

```python
processed = processed.apply(lambda x: " ".join(ps.stem(term) for term in x.split()))
```

```python
processed
```

```python
from nltk.tokenize import word_tokenize
```

```python
#crating a bag of words
all_words = []

for message in processed:
    words = word_tokenize(message)
    for word in words:
        all_words.append(word)
    
```

```python
all_words
```

```python
all_words = nltk.FreqDist(all_words)
```

```python
word_features = list(all_words.keys())[:1500]
```

```python
def find_features(message):
    words = word_tokenize(message)
    features = {}
    for word in word_features:
        features[word] = (word in words)
    return features
```

 ## 3. Scikit Learning Classifiers with NLTK

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
#Define a model to train
names = ['K Nearest Neighbour','Decsion Tree','Random Forest', 'Logistic Regression', 'SGD Classifier', 'Naive Bayes', 'SVM Linear']

classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    LogisticRegression(),
    SGDClassifier(max_iter = 100),
    MultinomialNB(),
    SVC(kernel = 'linear')
    
]

models = zip(names, classifiers)
```

```python
print(models)
```

```python
#wrap models in NLTK
from nltk.classify.scikitlearn import SklearnClassifier
```

```python
for name, model in models:
    nltk.model = SklearnClassifier(model)
    nltk.model.train(training)
    accuracy = nltk.classify.accuracy(nltk.model, testing) * 100
    print("{} Accuracy : {}".format(name,accuracy))
```

```python
# According to Results SVM is better.
```

```python
from sklearn.ensemble import VotingClassifier
names = ['K Nearest Neighbour','Decsion Tree','Random Forest', 'Logistic Regression', 'SGD Classifier', 'Naive Bayes', 'SVM Linear']

classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    LogisticRegression(),
    SGDClassifier(max_iter = 100),
    MultinomialNB(),
    SVC(kernel = 'linear')
    
]

models = zip(names, classifiers)
```

```python
nltk_ensemble = SklearnClassifier(VotingClassifier(estimators=list(models), voting = 'hard', n_jobs = -1))
nltk_ensemble.train(training)
accuracy = nltk.classify.accuracy(nltk_ensemble, testing) *100
print("Ensemble Method : Accuracy {}".format(accuracy))
```

```python
test_features, labels = zip(*testing)
predictions = nltk_ensemble.classify_many(test_features)
```

## 4. Classification Report and The Confusion Matrix

```python
print(classification_report(labels,predictions))
pd.DataFrame(
    confusion_matrix(labels, predictions),
    index = [['actual','actual'],['ham(real)','spam']],
    columns = [['predicted','predicted'],['ham(real)','spam']]
)
```
